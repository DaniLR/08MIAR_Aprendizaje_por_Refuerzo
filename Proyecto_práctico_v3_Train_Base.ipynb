````python
from __future__ import division

from PIL import Image
import numpy as np
import gym

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, InputLayer, Permute

if IN_COLAB:
  from tensorflow.keras.optimizers.legacy import Adam
else:
  from tensorflow.keras.optimizers.legacy import Adam

import tensorflow.keras.backend as K

from rl.agents.dqn import DQNAgent
from rl.policy import LinearAnnealedPolicy, BoltzmannQPolicy, EpsGreedyQPolicy
from rl.memory import SequentialMemory
from rl.core import Processor
from rl.callbacks import FileLogger, ModelIntervalCheckpoint
````

#### Configuración base

````python
# Formato de las imágenes de entrada
INPUT_SHAPE = (84, 84)
WINDOW_LENGTH = 4

# Configuración del entorno y extracción del número de acciones
env_name = 'SpaceInvaders-v0'
env = gym.make(env_name)
np.random.seed(123)
env.seed(123)
nb_actions = env.action_space.n
````

````python
print("Numero de acciones disponibles:" + str(nb_actions))

print("Formato de las observaciones:")
env.observation_space
````

````python
class AtariProcessor(Processor):
    def process_observation(self, observation):
        assert observation.ndim == 3  # (height, width, channel)
        img = Image.fromarray(observation)
        img = img.resize(INPUT_SHAPE).convert('L')
        processed_observation = np.array(img)
        assert processed_observation.shape == INPUT_SHAPE
        return processed_observation.astype('uint8')

    def process_state_batch(self, batch):
        processed_batch = batch.astype('float32') / 255.
        return processed_batch

    def process_reward(self, reward):
        return np.clip(reward, -1., 1.)
````

1. Implementación de la red neuronal

````python
# Formato de los datos de entrada
input_shape = (WINDOW_LENGTH,) + INPUT_SHAPE

# Definición del modelo
model = Sequential(name="DQN_SpaceInvaders")

# Datos de entrada
model.add(InputLayer(input_shape=input_shape, name="input_frames"))

# Compatibilidad de la entrda
model.add(Permute((2, 3, 1), name="permute_channels"))

# Capas convolucionales
model.add(Conv2D(32, (8, 8), strides=(4, 4), activation="relu", name="conv1"))
model.add(Conv2D(64, (4, 4), strides=(2, 2), activation="relu", name="conv2"))
model.add(Conv2D(64, (3, 3), strides=(1, 1), activation="relu", name="conv3"))

# Capa Flatten
model.add(Flatten(name="flatten"))

# Capas Densas
model.add(Dense(512, activation="relu", name="dense_hidden"))
model.add(Dense(nb_actions, activation="linear", name="output_values"))

# Imprimir información del modelo
print(model.summary())
````

2. Implementación de la solución DQN

````python
# Configuración de la memoria del algoritmo:
#    limit:         número máximo de transacciones
#    window_length: fotogramas consecutivos utilizados para formar un estado
memory = SequentialMemory(limit=1000000, window_length=WINDOW_LENGTH)
# Procesado de imágenes: convertir imagenes a escala de grises y normalización de los píxeles
processor = AtariProcessor()

# Estrategia de exploración:
#    EpsGreedyQPolicy: estrategia basada en epsilon
#    attr:             atributo epsilon
#    value_max:        valor inicial (máximo)
#    value_min:        valor mínimo (final) con el tiempo irá disminuyendo hasta este valor
#    value_test:       valor eps para el modo de prueba
#    nb_steps:         número de pasos, la disminución de eps será gradual durante este número de pasos
policy = LinearAnnealedPolicy(EpsGreedyQPolicy(), attr='eps',
                              value_max=1., value_min=.1, value_test=.05,
                              nb_steps=2000000)

# Configuración del agente DQN
#    model:               red neuronal
#    nb_actions:          número de accions del juego
#    policy:              estrategia de exploración
#    memory:              memoria del algoritmo
#    processor:           procesado de imágenes
#    nb_steps_warmup:     número de pasos en los que el algoritmo no entrena, solo juega para llenar la memoria con experiencias iniciales
#    gamma:               factor de actulización las recompensas
#    target_model_update: número de pasos hasta actualizar los valores de la red de predicción a la red de target
#    train_interval:      la red se entrena intervalo de fotogramas para optimización y alineamiento con los fotogramas del estado
dqn = DQNAgent(model=model, nb_actions=nb_actions, policy=policy,
               memory=memory, processor=processor,
               nb_steps_warmup=50000, gamma=.99,
               target_model_update=10000,
               train_interval=4)

# Compilación del agente
#    Adam:          optimizador Adam
#    learning_rate: tasa de aprendizaje
#    metrics:       métrica MAE (Error Absoluto Medio)
dqn.compile(Adam(learning_rate=.00025), metrics=['mae'])
````

````python
# Training part
weights_filename = 'dqn_{}_weights.h5f'.format(env_name)
checkpoint_weights_filename = 'dqn_' + env_name + '_weights_{step}.h5f'
log_filename = 'dqn_{}_log.json'.format(env_name)
callbacks = [ModelIntervalCheckpoint(checkpoint_weights_filename, interval=100000)]
callbacks += [FileLogger(log_filename, interval=100)]

# Comenzar el entrenamiento
dqn.fit(env, callbacks=callbacks, nb_steps=2000000, log_interval=10000, visualize=False)
#dqn.fit(env, callbacks=callbacks, nb_steps=20000, log_interval=100, visualize=False)

dqn.save_weights(weights_filename, overwrite=True)
````

````python
# Testing part to calculate the mean reward
weights_filename = 'dqn_{}_weights.h5f'.format(env_name)
dqn.load_weights(weights_filename)
dqn.test(env, nb_episodes=10, visualize=False)
````

3. Justificación de los parámetros seleccionados y de los resultados obtenidos

Propuesta 1: Double DQN (DDQN)

Implementación: Basta con instanciar el agente cambiando el parámetro nativo: enable_double_dqn=True.

Justificación teórica: Mitiga significativamente la inestabilidad en las estimaciones de recompensa, logrando curvas de aprendizaje más suaves y un rendimiento máximo superior en entornos con acciones discretas complejas como Space Invaders.

````python
# --- PROPUESTA 1: DOUBLE DQN ---
print("Inicializando configuración para Propuesta 1: Double DQN...")

# 1. Re-instanciar Entorno y Red Base
env = gym.make(ENV_NAME)
env.seed(123)

input_shape = (WINDOW_LENGTH,) + INPUT_SHAPE
model_p1 = Sequential()
if K.image_data_format() == 'channels_last':
    model_p1.add(Permute((2, 3, 1), input_shape=input_shape))
else:
    model_p1.add(Permute((1, 2, 3), input_shape=input_shape))

model_p1.add(Convolution2D(32, (8, 8), strides=(4, 4), activation='relu'))
model_p1.add(Convolution2D(64, (4, 4), strides=(2, 2), activation='relu'))
model_p1.add(Convolution2D(64, (3, 3), strides=(1, 1), activation='relu'))
model_p1.add(Flatten())
model_p1.add(Dense(512, activation='relu'))
model_p1.add(Dense(nb_actions, activation='linear'))

# 2. Configurar Memoria y Política de Exploración
memory_p1 = SequentialMemory(limit=1000000, window_length=WINDOW_LENGTH)
processor_p1 = AtariProcessor()
policy_p1 = LinearAnnealedPolicy(EpsGreedyQPolicy(), attr='eps', value_max=1., value_min=.1, value_test=.05, nb_steps=2000000)

# 3. Instanciar Agente ACTIVANDO Double DQN
dqn_p1 = DQNAgent(model=model_p1, nb_actions=nb_actions, policy=policy_p1,
                  memory=memory_p1, processor=processor_p1,
                  nb_steps_warmup=50000, gamma=.99,
                  target_model_update=10000, train_interval=4,
                  enable_double_dqn=True) # <-- MEJORA AQUI

dqn_p1.compile(Adam(learning_rate=.00025), metrics=['mae'])

# 4. Callbacks con rutas únicas a Google Drive
weights_final_p1 = '/content/drive/MyDrive/dqn_spaceinvaders_final_p1.h5f'
checkpoint_p1 = '/content/drive/MyDrive/dqn_spaceinvaders_p1_{step}.h5f'
log_p1 = '/content/drive/MyDrive/dqn_spaceinvaders_log_p1.json'

callbacks_p1 = [ModelIntervalCheckpoint(checkpoint_p1, interval=100000), FileLogger(log_p1, interval=100)]

# 5. Lanzar Entrenamiento
# dqn_p1.fit(env, callbacks=callbacks_p1, nb_steps=2000000, log_interval=10000, visualize=False)
# dqn_p1.save_weights(weights_final_p1, overwrite=True)
````

Propuesta 2: Modificación de la Frecuencia y Estilo de Actualización del Target

Implementación: Configurar target_model_ update=1e-3 (o 1e-2) al instanciar el DQNAgent.

Justificación teórica: En lugar de mantener congelada la red objetivo y cambiar sus pesos abruptamente cada 10,000 pasos, una actualización suave incorpora en cada paso un pequeño porcentaje térmico ($\tau$) de los pesos actuales ($\theta_{target} \leftarrow \tau \theta_{online} + (1-\tau)\theta_{target}$). Esto evita oscilaciones destructivas en el aprendizaje y ayuda a estabilizar la política del agente en Space Invaders, donde perder una vida cambia bruscamente el mapa de características visuales.

````python
# --- PROPUESTA 2: SOFT UPDATE ---
print("Inicializando configuración para Propuesta 2: Soft Update...")

# 1. Re-instanciar Entorno y Red Base
env = gym.make(ENV_NAME)
env.seed(123)

input_shape = (WINDOW_LENGTH,) + INPUT_SHAPE
model_p2 = Sequential()
if K.image_data_format() == 'channels_last':
    model_p2.add(Permute((2, 3, 1), input_shape=input_shape))
else:
    model_p2.add(Permute((1, 2, 3), input_shape=input_shape))

model_p2.add(Convolution2D(32, (8, 8), strides=(4, 4), activation='relu'))
model_p2.add(Convolution2D(64, (4, 4), strides=(2, 2), activation='relu'))
model_p2.add(Convolution2D(64, (3, 3), strides=(1, 1), activation='relu'))
model_p2.add(Flatten())
model_p2.add(Dense(512, activation='relu'))
model_p2.add(Dense(nb_actions, activation='linear'))

# 2. Configurar Memoria y Política de Exploración
memory_p2 = SequentialMemory(limit=1000000, window_length=WINDOW_LENGTH)
processor_p2 = AtariProcessor()
policy_p2 = LinearAnnealedPolicy(EpsGreedyQPolicy(), attr='eps', value_max=1., value_min=.1, value_test=.05, nb_steps=2000000)

# 3. Instanciar Agente aplicando SOFT UPDATE
dqn_p2 = DQNAgent(model=model_p2, nb_actions=nb_actions, policy=policy_p2,
                  memory=memory_p2, processor=processor_p2,
                  nb_steps_warmup=50000, gamma=.99,
                  target_model_update=1e-3, train_interval=4) # <-- MEJORA AQUI (Valor < 1 activa Soft Update)

dqn_p2.compile(Adam(learning_rate=.00025), metrics=['mae'])

# 4. Callbacks con rutas únicas a Google Drive
weights_final_p2 = '/content/drive/MyDrive/dqn_spaceinvaders_final_p2.h5f'
checkpoint_p2 = '/content/drive/MyDrive/dqn_spaceinvaders_p2_{step}.h5f'
log_p2 = '/content/drive/MyDrive/dqn_spaceinvaders_log_p2.json'

callbacks_p2 = [ModelIntervalCheckpoint(checkpoint_p2, interval=100000), FileLogger(log_p2, interval=100)]

# 5. Lanzar Entrenamiento (Descomenta la línea de abajo para empezar)
# dqn_p2.fit(env, callbacks=callbacks_p2, nb_steps=2000000, log_interval=10000, visualize=False)
# dqn_p2.save_weights(weights_final_p2, overwrite=True)
````

Propuesta 3: Incremento de la Capacidad de Representación en la Red Neuronal (Cambio de Arquitectura)

El juego Space Invaders presenta dinámicas espaciotemporales complejas (múltiples filas de alienígenas moviéndose en bloque, proyectiles rápidos en trayectorias descendentes y barreras que se destruyen). La capa densa original de 512 neuronas puede quedarse corta para generalizar eficazmente todas estas combinaciones.

Implementación: Modificar la sección del modelo añadiendo una segunda capa densa o expandiendo la actual:

````python
# --- PROPUESTA 3: CAMBIO DE ARQUITECTURA ---
print("Inicializando configuración para Propuesta 3: Arquitectura Ampliada...")

# 1. Re-instanciar Entorno
env = gym.make(ENV_NAME)
env.seed(123)

# 2. CONSTRUCCIÓN DE LA NUEVA RED AMPLIADA
input_shape = (WINDOW_LENGTH,) + INPUT_SHAPE
model_p3 = Sequential()
if K.image_data_format() == 'channels_last':
    model_p3.add(Permute((2, 3, 1), input_shape=input_shape))
else:
    model_p3.add(Permute((1, 2, 3), input_shape=input_shape))

# Mantenemos las convolucionales para extraer características visuales
model_p3.add(Convolution2D(32, (8, 8), strides=(4, 4), activation='relu'))
model_p3.add(Convolution2D(64, (4, 4), strides=(2, 2), activation='relu'))
model_p3.add(Convolution2D(64, (3, 3), strides=(1, 1), activation='relu'))
model_p3.add(Flatten())

# MEJORA: Bloque denso expandido con más neuronas y profundidad
model_p3.add(Dense(1024, activation='relu')) # <-- Nueva capa de alta capacidad
model_p3.add(Dense(512, activation='relu'))  # <-- Capa de consolidación
model_p3.add(Dense(nb_actions, activation='linear'))

print(model_p3.summary()) # Para verificar la nueva estructura en el notebook

# 3. Configurar Memoria y Política de Exploración
memory_p3 = SequentialMemory(limit=1000000, window_length=WINDOW_LENGTH)
processor_p3 = AtariProcessor()
policy_p3 = LinearAnnealedPolicy(EpsGreedyQPolicy(), attr='eps', value_max=1., value_min=.1, value_test=.05, nb_steps=2000000)

# 4. Instanciar Agente DQN estándar pero con el nuevo modelo model_p3
dqn_p3 = DQNAgent(model=model_p3, nb_actions=nb_actions, policy=policy_p3,
                  memory=memory_p3, processor=processor_p3,
                  nb_steps_warmup=50000, gamma=.99,
                  target_model_update=10000, train_interval=4)

dqn_p3.compile(Adam(learning_rate=.00025), metrics=['mae'])

# 5. Callbacks con rutas únicas a Google Drive
weights_final_p3 = '/content/drive/MyDrive/dqn_spaceinvaders_final_p3.h5f'
checkpoint_p3 = '/content/drive/MyDrive/dqn_spaceinvaders_p3_{step}.h5f'
log_p3 = '/content/drive/MyDrive/dqn_spaceinvaders_log_p3.json'

callbacks_p3 = [ModelIntervalCheckpoint(checkpoint_p3, interval=100000), FileLogger(log_p3, interval=100)]

# 6. Lanzar Entrenamiento (Descomenta la línea de abajo para empezar)
# dqn_p3.fit(env, callbacks=callbacks_p3, nb_steps=2000000, log_interval=10000, visualize=False)
# dqn_p3.save_weights(weights_final_p3, overwrite=True)
````
